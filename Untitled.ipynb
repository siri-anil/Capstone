```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
hotness=pd.read_csv('data/RDC_Inventory_Hotness_Metrics_Metro_History .csv',nrows=24000)
hotness
```

```python
hotness.columns
```

```python
hotness_df=hotness[['month_date_yyyymm', 'cbsa_code', 'cbsa_title','nielsen_hh_rank',
       'hotness_rank', 'hotness_score',
       'supply_score', 'demand_score', 'median_days_on_market',
       'median_days_on_market_mm', 'median_dom_mm_day',
       'median_days_on_market_yy', 'median_dom_yy_day', 'median_listing_price',
       'median_listing_price_mm', 'median_listing_price_yy']]
```

```python
hotness_df['month_date_yyyymm']=pd.to_datetime(hotness_df['month_date_yyyymm'],format='%Y%m')
hotness_df
```

```python
hotness_df['year']=hotness_df['month_date_yyyymm'].dt.year
hotness_df
```

```python
hotness_df['state_id']=hotness_df['cbsa_title'].str[-2:]
hotness_df
```

```python
sales_hotness_rank=hotness_df.groupby(['state_id','year'])['hotness_rank'].mean().reset_index()
sales_hotness_rank
```
